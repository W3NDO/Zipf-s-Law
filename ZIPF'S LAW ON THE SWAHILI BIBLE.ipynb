## **Zipf's Law:**
According to Encyclopedia Britanicca, Zipf's Law is the assertion that the frequency **f** of certain events are inversely proportional to their rank. It was proposed by American Linguist George Kingsley Zipf for the frequency of usage of different words in the English language; this frequency is given approximately by 

f(r) ≅ 0.1/r. 

Thus, the most common word (rank 1) in English, which is the, occurs about one-tenth of the time in a typical text; the next most common word (rank 2), which is of, occurs about one-twentieth of the time; and so forth. Another way of looking at this is that a rank r word occurs 1/r times as often as the most frequent word, so the rank 2 word occurs half as often as the rank 1 word, the rank 3 word one-third as often, the rank 4 word one-fourth as often, and so forth. Beyond about rank 1,000, the law completely breaks down. Zipf’s law purportedly has been observed for many other statistics that follow an exponential distribution.

It has also been observed that Zipf's law does not only apply to the English language but all languages. For instance if we look at the analysis of words in an exerpt of the Swahili bible, we observe the following:

```python
import numpy as np
import matplotlib.pyplot as plt


exerpt = open('swahili bible.txt', 'r')
wordDictionary = {}
wordCount = 0
for line in exerpt:
    wordArray = line.split(' ')
    for word in wordArray:
        wordCount+=1
        if word in wordDictionary:
            wordDictionary[word] += 1
        else:
            wordDictionary[word] = 1
            
wordDictionary = sorted(wordDictionary.items(), key=lambda x: x[1], reverse=True) 
#sort the dictionary from largest to smallest
print("The 10 most frequent words in the swahili bible are: ")
for i in range(10):
    print(str(i+1) + ':' + str(wordDictionary[i]))

y_Value =[]
for key in wordDictionary:
    y_Value.append(key[1])
    
y_Value = np.array(y_Value)
x_Value = []
for i in y_Value:
    x_Value.append(wordCount * (1/i))
   
    
x= np.log(x_Value[:200])
y = np.log(y_Value[:200])
plt.plot(x,y, c="r")
plt.title("Log Log Scale")
plt.xlabel('Word Rank')
plt.ylabel("Word Frequency")
plt.grid(True)
plt.show()

x= x_Value[:200]
y = y_Value[:200]
plt.scatter(x,y, c='r')
plt.title("Normal")
plt.xlabel('Word Rank')
plt.ylabel("Word Frequency")
plt.grid(True)
plt.show()
```

The above graphs display the general power law characteristic to Zipf's law. 

The most common word, 'na' occurs 5952 times and the second most common word, 'ya', occurs 3638 times. According to zipf's law, the second word should occur about half as often as the first word. Half of 5952 is 2976. So clearly it isn't an exact law. However, this disparity can be attributed to the size of the data set I used. Surely if I used a bigger data set, I would end up with a more precise figure. 

Regardless, we are still trying to understand why Zipf's law occurs the way it does, but it is surely an interesting occurrence.
